### PIE CHART

```python
Beginner= data.loc[data['level'] == 'Beginner Level'].count()[0]
Intermediate= data.loc[data['level'] == 'Intermediate Level'].count()[0]
All = data.loc[data['level'] == 'All Levels'].count()[0]
Expert = data.loc[data['level'] == 'Expert Level'].count()[0]

labels = ['Beginner Level', 'Intermediate Level ','All Levels','Expert Level']
colors = ['blue','red','orange','cyan']

plt.pie([Beginner,Intermediate,All,Expert], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('ENCHANCING UDEMY WITH DATA ANALYSIS')

plt.show()
```
